```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 不要な警告を非表示にする
import warnings
warnings.filterwarnings('ignore')
```

```python
# bostonデータを格納したクラス
from sklearn.datasets import load_boston

# インスタンスを生成
boston = load_boston()
```

```python
#説明変数の確認
columns = boston.feature_names
columns
```

```python
#boston.dataに説明変数がboston.targetに目的変数が格納されているので確認

boston.data
```

```python
boston.target[:10]
```

```python
#PandasのDataFrameに変換
df_data = pd.DataFrame(boston.data, columns=boston.feature_names)
df_data.head()
```

```python
#目的変数も同様　変数名はMEDV(中央値)
df_target = pd.DataFrame(boston.target, columns=['MEDV'])
df_target.head()
```

```python
# 説明変数と目的変数のDataFrameを結合する
df = pd.concat([df_data, df_target], axis=1)
df.head()

#axis=1...横
```

```python
#相関係数の確認
df.corr()
```

```python
#基本統計量の確認
df.describe()
```

```python
# 説明変数のデータを用意
#今回は相関が強い２つの変数を採用
X = df.loc[:, ['LSTAT', 'RM']].values
X
```

```python
# 教師データを用意
y = df.loc[:, ['MEDV']].values

# 最初の１０個を表示して確認
y[:10]
```

```python
# ホールド・アウト法によるデータの分割
# test_sizeで分割する割合 random_stateで分割の際のブレ？を作れる
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# 重回帰の線形回帰を実行
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# 重回帰線形モデルの学習
lr.fit(X_train, y_train)
```

```python
# 定数項を出力
lr.intercept_
```

```python
# 偏回帰係数を出力
# [LSTATの係数, RMの係数]
lr.coef_
```
